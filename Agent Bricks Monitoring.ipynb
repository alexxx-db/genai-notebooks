This gets information about a served entity from the workspace

```sql
SELECT * 
FROM system.serving.served_entities
```

```sql
SELECT * 
FROM system.serving.served_entities
WHERE endpoint_name = :served_entity_name
```

Currently, you can't see endpoint usage for agent bricks, which makes tracking things like 403/429 difficult. It is on the roadmap!

```sql
SELECT *
FROM system.serving.endpoint_usage
WHERE served_entity_id IN (
  SELECT served_entity_id 
  FROM system.serving.served_entities
  WHERE endpoint_name = :served_entity_name
)
```

We can see billing via served entity name though.

```sql
SELECT *
FROM system.billing.usage
WHERE usage_metadata.endpoint_name = :served_entity_name
```

```sql
SELECT usage_date, SUM(usage_quantity) as daily_dbu
FROM system.billing.usage
WHERE usage_metadata.endpoint_name = :served_entity_name
AND usage_unit = 'DBU'
GROUP BY usage_date
ORDER BY usage_date DESC
```
